```python
import pandas as pd
from pathlib import Path
```

```python
root = Path('./data')
def create_path(root, journal):
    suffix = '_stats_per_dots.csv'
    return root.joinpath(journal).joinpath(journal+suffix)

df = pd.read_csv(create_path(root,'gazzetta_dello_sport_3plus'))
c2k = ['code', 'dots', 'giocatore', 'squadra', 'ruolo', 'yyss']

df = df[c2k]
```

```python
df['giocatore'] = df['giocatore'].str.rstrip().str.lstrip().str.lower()
df['squadra'] = df['squadra'].str.rstrip().str.lstrip().str.lower()

df['last_len'] = df['giocatore'].str.split(' ').str[-1].str.len()
df['spaces'] = df['giocatore'].str.count(' ')

df[df.yyss=='2009-10']
```
