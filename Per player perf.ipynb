```python
import pandas as pd
import plotly.graph_objects as go
from plotly.graph_objs.scatter.marker import Line
import plotly.express as px
import itertools
import numpy as np
from sklearn.linear_model import LinearRegression
 
pd.set_option('display.max_rows', 500)

data_path = './data/stats.csv'
df = pd.read_csv(data_path)
df['time_index'] = df['dots'].astype(str) +'--'+ df['yyss'].astype(str)
df['giocatore'] = df.giocatore.str.lower()
df.columns
```

```python
# def myplot(giocatore, df):
giocatore='insigne'


player = df[df.giocatore.str.contains(giocatore)].reset_index()
lists = [df.yyss.unique().tolist(), list(range(1,39))]
time_df = pd.DataFrame(index=[str(val[1])+'--'+str(val[0]) for val in itertools.product(*lists)])
time_df['fantavoto'] = np.nan
player_complete = player.merge(time_df, left_on='time_index', right_index=True, how='outer', suffixes=('', '_y'))
player_complete['dots'] = player_complete['time_index'].str.split('--').str[0].astype(int)
player_complete['yyss'] = player_complete['time_index'].str.split('--').str[1]
player_complete = player_complete.sort_values(by=['yyss', 'dots']).reset_index().drop('index',axis=1)
player_complete['x'] = player_complete.index


title = "%s - %s" %(giocatore.capitalize(), player.ruolo[0])

player = player_complete.copy()
player['text'] = player['dots'].astype(str) + "<br>" +player['yyss'].str[2:]
fig = px.scatter(player, x=player.x, y=player.fantavoto, 
                 trendline="ols", title=title, hover_data=["dots", "yyss"])
fig.update_traces(hovertemplate='Giornata: %{dots} <br>Stagione: %{yyss} <br><b>voto: %{y}</b>')

not_nan = player.fillna(0).reset_index()
X = not_nan.index.values.reshape(-1, 1)

model = LinearRegression()
model.fit(X, not_nan.fantavoto)
x_range = np.linspace(X.min(), X.max(), 100)
y_range = model.predict(x_range.reshape(-1, 1))
fig.add_traces(go.Scatter(x=x_range, y=y_range, name='Regression Fit'))

fig.show()

#     return fig
```
